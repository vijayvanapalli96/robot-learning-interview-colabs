# 04 - Policy Evaluation and Failure Logging\n\nThis notebook creates a tiny evaluation log and turns it into interview-ready metrics.

```python
import pandas as pd\nimport matplotlib.pyplot as plt\n\nrows = [\n    {'trial': 1, 'success': 1, 'interventions': 0, 'latency_ms': 85, 'failure_mode': 'none'},\n    {'trial': 2, 'success': 0, 'interventions': 1, 'latency_ms': 110, 'failure_mode': 'contact_loss'},\n    {'trial': 3, 'success': 1, 'interventions': 0, 'latency_ms': 92, 'failure_mode': 'none'},\n    {'trial': 4, 'success': 0, 'interventions': 2, 'latency_ms': 140, 'failure_mode': 'yaw_oscillation'},\n    {'trial': 5, 'success': 0, 'interventions': 1, 'latency_ms': 125, 'failure_mode': 'localization_drop'},\n    {'trial': 6, 'success': 1, 'interventions': 0, 'latency_ms': 88, 'failure_mode': 'none'},\n]\ndf = pd.DataFrame(rows)\ndisplay(df)\n\nprint('success rate:', df.success.mean())\nprint('mean interventions:', df.interventions.mean())\nprint('p95-ish latency:', df.latency_ms.quantile(0.95))\nprint('\nfailure modes')\nprint(df.failure_mode.value_counts())
```

```python
df.failure_mode.value_counts().plot(kind='bar', title='Failure mode distribution')\nplt.ylabel('count')\nplt.show()
```
