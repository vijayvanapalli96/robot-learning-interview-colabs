# 02 - Action Chunking and Latency Tradeoffs\n\nThis notebook simulates why chunk size affects smoothness and responsiveness.

```python
import numpy as np\nimport matplotlib.pyplot as plt\n\nnp.random.seed(7)\nT = 120\ntarget = np.sin(np.linspace(0, 4*np.pi, T))\ntarget[60:] += 0.8  # sudden new observation / goal change\n\ndef chunked_policy(target, chunk):\n    actions = []\n    for t in range(0, len(target), chunk):\n        goal = target[t]\n        plan = np.linspace(goal, goal, chunk) + np.random.normal(0, 0.03, chunk)\n        actions.extend(plan)\n    return np.array(actions[:len(target)])\n\nplt.figure(figsize=(10, 4))\nplt.plot(target, label='changing target', linewidth=3)\nfor chunk in [1, 4, 12, 24]:\n    plt.plot(chunked_policy(target, chunk), label=f'chunk={chunk}', alpha=0.8)\nplt.axvline(60, color='black', linestyle='--', linewidth=1)\nplt.title('Small chunks react quickly; large chunks are smoother but stale longer')\nplt.legend()\nplt.show()
```

## What to say\n\n> A larger action chunk can produce smoother behavior, but it can also make the robot less reactive to new observations. A smaller chunk can react faster, but may create jitter or discontinuities. In robotics, latency and action representation change the behavior of the robot.
